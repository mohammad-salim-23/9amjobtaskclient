```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

##**histogram using Seaborn**

```python
#figure level
sns.displot(kind='hist',data=student,x='attendance_rate',col='gender')
```
